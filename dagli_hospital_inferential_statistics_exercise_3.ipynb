# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

# 1 .Setup an appropriate hypothesis test.
I will check whether there is a significant difference between the means of hospitals/facilties with small capacity (< 300) and hospitals/facilties with small capacity (>= 300). Therefore my hypothesis is:

H0: There is no difference between hospital with small capacity (<300) and hospitals with large capacity (>=300).
H1: There is difference between hospital with small capacity and hospitals with large capacity.


# 2. Compute and report the observed significance value (or p-value).
I would use t test to determine whether or not there is difference between the groups. 

```python
#first I will divide the datasets two subsets based on the capacity (<300)
ndlow_300=df[df['Number of Discharges']<300] # subdataframe with number of discharges < 300
print(ndlow_300['Excess Readmission Ratio'].mean()) # Calculation mean
print(len(ndlow_300[ndlow_300['Excess Readmission Ratio']>1])/(len(ndlow_300))) # Rate of readmission rate greater than 1
```

```python
#first I will divide the datasets two subsets based on the capacity (>=300)
ndhigh_300=df[df['Number of Discharges']>=300] # subdataframe with number of discharges >= 300
print(ndhigh_300['Excess Readmission Ratio'].mean()) # Calculation mean
print(len(ndhigh_300[ndhigh_300['Excess Readmission Ratio']>1])/(len(ndhigh_300))) # Rate of readmission rate greater than 1
```

# 3. Report statistical significance for  α= .01.
As seen below, p value is less than .01 so that we can conclude that there is a statistically significant difference between two groups.

# 4. Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?
However, statistical significance does not necessarily mean that the results are practically significant in a real-world. Large sample size may produce small p vlaues, which lead to statistically siginificant result. Also, low sample variability may produce very small p values. These two factors may result in statistically significant but trivial. To assess the practical significance, I will check the confidence interval as well as calculate the effect size (Cohen's d) to see how large the effect is between two group means. 


```python
# Run independent t-test
from scipy import stats
ind_t_test=stats.ttest_ind(ndlow_300['Excess Readmission Ratio'],ndhigh_300['Excess Readmission Ratio'])
print('The results of the independent t-test are: \n\tt-value = {:4.3f}\n\tp-value = {:4.3f}'.format(ind_t_test[0],ind_t_test[1]))
```

```python
#Calculating confidence interval for both groups

ndlow_SE= np.std(ndlow_300['Excess Readmission Ratio'])/np.sqrt(len(ndlow_300['Excess Readmission Ratio'])) # Standard Error
ci=2.58 # 99% confidence interval z score
ndlow_ci_low=np.mean(ndlow_300['Excess Readmission Ratio']) - ci * ndlow_SE
ndlow_ci_high=np.mean(ndlow_300['Excess Readmission Ratio']) + ci * ndlow_SE
print('Confidence interval for hospitals with small capacity:', ndlow_ci_low, '-', ndlow_ci_high)

ndhigh_SE= np.std(ndhigh_300['Excess Readmission Ratio'])/np.sqrt(len(ndhigh_300['Excess Readmission Ratio'])) # Standard Error
ndhigh_ci_low=np.mean(ndhigh_300['Excess Readmission Ratio']) - ci * ndhigh_SE
ndhigh_ci_high=np.mean(ndhigh_300['Excess Readmission Ratio']) + ci * ndhigh_SE
print('Confidence interval for hospitals with large capacity:', ndhigh_ci_low, '-', ndhigh_ci_high)
```

Confidence Intervals describe how uncertain our estimated difference is. CIs provide an estimate of what this difference is. A narrower CI will indicate a more precise estimate, while a wider CI indicates a less precise estimate. As we checked the CIs found above the estimated differences for both groups is very close to each other. Also, the meanx (0.99 and 1.01) and standard deviations (0.10 and 0.08) of each group are almost same. It seems there is very weak practical significance between the groups.


```python
# For hospitals with small capacity, Draw some random sample with replacement and append mean to mean_temp.
ndlow, sims = [], 100000
for i in range(sims):
    call_sample = np.random.choice(ndlow_300['Excess Readmission Ratio'], replace=True, size=len(ndlow_300['Excess Readmission Ratio']))
    sample_mean = call_sample.mean()
    ndlow.append(sample_mean)
    
# Calculate bootstrapped mean and 99% confidence interval.
boot_mean = np.mean(ndlow)
boot_99_ci = np.percentile(ndlow, [0.5, 99.5])
print("Bootstrapped hospitals with small capacity = {}, 99% CI = {}".format(boot_mean, boot_99_ci))
```

```python
#Bootstrapping 

ndhigh, sims = [], 100000
for i in range(sims):
    call_sample = np.random.choice(ndhigh_300['Excess Readmission Ratio'], replace=True, size=len(ndhigh_300['Excess Readmission Ratio']))
    sample_mean = call_sample.mean()
    ndhigh.append(sample_mean)
    
# Calculate bootstrapped mean and 95% confidence interval.
boot_mean = np.mean(ndhigh)
boot_99_ci = np.percentile(ndhigh, [0.5, 99.5])
print("Bootstrapped hospitals with small capacity = {}, 99% CI = {}".format(boot_mean, boot_99_ci))
```

```python
# calculate the Cohen's d between two groups.
from numpy.random import randn
from numpy.random import seed
from numpy import mean
from numpy import var
from math import sqrt
 
# function to calculate Cohen's d for independent samples
def cohend(d1, d2):
    # calculate the size of samples
    n1, n2 = len(d1), len(d2)
    # calculate the variance of the samples
    s1, s2 = var(d1, ddof=1), var(d2, ddof=1)
    # calculate the pooled standard deviation
    s = sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
    # calculate the means of the samples
    u1, u2 = mean(d1), mean(d2)
    # calculate the effect size
    return (u1 - u2) / s
 

# calculate cohen's d
d = cohend(ndlow_300['Excess Readmission Ratio'], ndhigh_300['Excess Readmission Ratio'])
print('Cohens d: %.3f' % d)
```

 Cohen suggested that d=0.2 be considered a 'small' effect size, 0.5 represents a 'medium' effect size and 0.8 a 'large' effect size. As I found the Cohen's d (0.157) less than 0.2, I lean to make a conclusion that two group's mean don't differ by 0.2 standard deviations or more the difference between the groups' mean is not important. However, experts in the field of interests (public health maybe) should decide whether or not the results have any practical significance.
